### Correlation heatmap

```python

data = pd.read_csv('/Users/Yu/Documents/GitHub/Midterm_Project/data/cleaned_data.csv')

# Select columns for correlation matrix
selected_columns = ['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Generosity']

# Compute correlation matrix
correlation_matrix = data[['Happiness Score'] + selected_columns].corr()

# Plot correlation heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='copper', center=0)
plt.title('Correlation Heatmap for Canada')
plt.show()
```

### Distribution of all numeric data

```python
# Exclude 'Year' and 'Happiness Rank' columns
columns_to_exclude = ['Year', 'Happiness Rank']
columns_to_plot = [col for col in data.columns if col not in columns_to_exclude]

plt.rcParams['figure.figsize'] = (12, 12)

data[columns_to_plot].hist()

plt.tight_layout()
plt.show()
```
